## Preprocessing

## Feature Scaling

```python
s=StandardScaler()
```

```python
iris
```

```python
x = iris.drop(["Species"],axis=1)
```

```python
y = iris["Species"]
```

```python
x
```

```python
y
```

```python
scaler = s.fit(x)
```

```python
X_stdscaler = scaler.transform (x)
```

```python
X_stdscaler
```

## Label Encoder

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
y
```

## Splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5,random_state=62)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Model Selection

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = "linear")
print(svc.fit(x_train,y_train))
```

```python
y_predict = svc.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score 
```

```python
accuracy_score(y_predict,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
print(knn.fit(x_train,y_train))
```

```python
y_predict = knn.predict(x_test)
```

```python
accuracy_score(y_predict,y_test)
```

```python
print("Accuracy ofthe give Dataset is :",accuracy_score(y_predict,y_test)*100,'%')
```
